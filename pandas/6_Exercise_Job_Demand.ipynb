```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_poland = df[df['job_country'] == 'Poland'].copy()
df_poland
```

```python
df_poland['job_posted_month'] = df_poland['job_posted_date'].dt.strftime('%B')
df_poland
```

```python
df_poland_pivot = df_poland.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size',
)
df_poland_pivot
```

```python
df_poland_pivot.reset_index(inplace=True)
```

```python
df_poland_pivot['month_no'] = pd.to_datetime(df_poland_pivot['job_posted_month'], format='%B').dt.month
df_poland_pivot['month_no']
```

```python
df_poland_pivot
```

```python
df_poland_pivot.sort_values('month_no', inplace=True)
df_poland_pivot.set_index('job_posted_month', inplace=True)
df_poland_pivot.drop(columns='month_no', inplace=True)
```

```python
df_poland_pivot
```

```python
df_poland_pivot.plot(kind='line')
```

```python
top_3 = df_poland['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

```python
df_poland_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
